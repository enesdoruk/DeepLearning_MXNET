## GRADIENT DESCENT

```python
import d2l 
from mxnet import *
from mxnet.gluon import nn
npx.set_np()
```

```python
def f(x):
    return x ** 2

def gradf(x):
    return 2 * x


def gd(eta):
    x = 10
    results = [x]
    for i in range(10):
        x -= eta * gradf(x)
        results.append(x)
    print('epoch 10, x', x)
    return results
```

```python
res = gd(0.2)
```

```python
def show_trace(res):
    n = max(abs(min(res)), abs(max(res)))
    f_line = np.arange(-n, n, 0.01)
    d2l.set_figsize((3.5, 2.5))
    d2l.plot([f_line, res], [[f(x) for x in f_line], [f(x) for x in res]], 
            'x', 'f(x)', fmts=['-', '-o'])

show_trace(res)
```

```python
show_trace(gd(0.05)) # with learning rate
```

```python
show_trace(gd(1.1))
```
